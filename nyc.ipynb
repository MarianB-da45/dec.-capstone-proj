```python
import pandas as pd
import requests
import sqlalchemy 
from sqlalchemy import create_engine
from sqlalchemy.orm import sessionmaker
from sqlalchemy import text
import os



```

```python
from dotenv import load_dotenv
from sqlalchemy import create_engine

load_dotenv(True)override=


```

```python
# Extract data from a CSV
AgencyMaster = pd.read_csv(r'Data\AgencyMaster.csv') 
EmpMaster = pd.read_csv(r'Data\EmpMaster.csv') 
nycpayroll_2020 = pd.read_csv(r'Data\nycpayroll_2020.csv') 
nycpayroll_2021 = pd.read_csv(r'Data\nycpayroll_2020.csv') 
TitleMaster = pd.read_csv(r'Data\TitleMaster.csv') 

```

```python
AgencyMaster.head()
```

```python
#Rename the 'AgencyCode' column to 'AgencyID'
nycpayroll_2021.rename(columns={'AgencyCode': 'AgencyID'},inplace=True)
```

```python
nycpayroll_2021.head()
```

```python
#concate as one dataframe
nycpayroll_df = pd.concat([nycpayroll_2020, nycpayroll_2021], ignore_index=True)

# Display the DataFrame
nycpayroll_df
```

```python
#Absolute command 
# Convert only numeric columns to positive values
numeric_cols = nycpayroll_df.select_dtypes(include=['number']).columns
nycpayroll_df[numeric_cols] = nycpayroll_df[numeric_cols].applymap(lambda x: abs(x) if x < 0 else x)

print("\nDataFrame after converting numeric columns only:")
print(nycpayroll_df)
```

```python
#check for null values
print("\nNull Values in DataFrame:")
print(nycpayroll_df.isnull())
```

```python
# Check if the DataFrame contains any null values
print("\nDoes the DataFrame contain any null values?")
print(nycpayroll_df.isnull().values.any())
```

```python
# Show rows with any null values
print("\nRows with Any Null Values:")
print(nycpayroll_df[nycpayroll_df.isnull().any(axis=1)])
```

```python
#


# Convert to datetime
nycpayroll_df['AgencyStartDate'] = pd.to_datetime(nycpayroll_df['AgencyStartDate'])

print("\nDataFrame with Datetime Format:")
print(nycpayroll_df)
```

```python
# writing combine dataframe to csv
df = pd.DataFrame(nycpayroll_df)

# Write the DataFrame to a CSV file
output_file = "nycpayroll_new.csv"
df.to_csv(output_file, index=False)  # index=False to exclude the index column

print(f"DataFrame has been written to {output_file}")
```

```python
## create an engine
load_dotenv('.env', override=True)
engine = create_engine("postgresql+psycopg2://{user}:{password}@{host}:{port}/{dbname}".format(
                           user = os.getenv('pg_user'),
                           password = os.getenv('pg_password'),
                           host = os.getenv('pg_host'),
                           port = os.getenv('pg_port'),
                           dbname = os.getenv('pg_database')     
                                )
                        )
    
```

```python
### CREATE TABLE
## SPECIFY THE REQUIRED COLUMNS

schema = 'staging_schema'

''' my_columns = {
    'FiscalYear' : int,
    'PayrollNumber' : int,
    'AgencyID' :  int,
    'AgencyName' : str,
    'EmployeeID' : int,
    'LastName' : str,
    'FirstName' : str,
    'AgencyStartDate' : datetime,
    'WorkLocationBorough' : str
} '''



# Load data into PostgreSQL
## Loading the data to postgresql

AgencyMaster.to_sql('AgencyMaster', schema=schema, con=engine, if_exists = 'replace', index=False)
EmpMaster.to_sql('EmpMaster', schema=schema, con=engine, if_exists = 'replace', index=False)
nycpayroll_df.to_sql('nycpayroll_df', schema=schema, con=engine, if_exists = 'replace',index=False)
TitleMaster.to_sql('TitleMaster', schema=schema, con=engine, if_exists = 'replace', index=False)
try:   
    print(f"Data successfully loaded into the table '{schema}'.")

except Exception as e:

    print(f"Error loading data: {e}")
```
